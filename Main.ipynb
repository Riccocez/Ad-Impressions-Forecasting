## 3. Prediction of ad impressions in Game X for February

We are interested in forecasting the total ad impressions in a Game X for February. We built a proof of concept of the forecasting for the total ad impressions in February. The main reason of this consideration is because of the small size of the dataset provided.

We assumed this task as a regression problem, and developed a deep learning model for modelling the Time Series dataset. We preferred this approach due to its predicting performance proved to be the best among other regression techniques such as Lasso or Ridge Regression. 

We will be using the given time series data and build a model with a simple architecture. In order to get a more complex and accurate solution, it would be very useful considering manipulating the data in various ways such as rescaling it or convert it into an stationary time series. By rescaling we could consider applying transformations like the square or cube root, in order to penalize dramatic values in data. Stationary data would be a very useful approach to remove some trend or seasonality when having a bigger dataset.

Below, we followed the next steps:
    - We split our data into a training and testing set
    - We trained different models starting from simpler to more complex regression techniques
    - We compared the results of each model
    - We identified some benefits and limitations of the models and in the dataset
    - Based on those visualizations and analysis, we selected the best possible solution
    - Then, we tunned our selected model to improve its forecasting performance
    - We saved in different files the forecasting results and fitting behavior of the selected model
    - Finally, we created data visualizations to explain the behavior found in the selected model

```python
raw_df = data.get_file(2)
df = manip.build_features(raw_df)
imp_df = manip.select_features(df,['Date','Impressions',
                             'dayofweek', 'dayofyear',
                             'weekofyear','Prev_Impressions'],indexfield='Date')
```

```python
# fix random seed for reproducibility
np.random.seed(7)

dataset = imp_df.values
dataset = dataset.astype('float32')

train,test,train_size = manip.split_train_test(dataset,size=0.67)
test_df = manip.get_test_data(imp_df,["dayofweek","dayofyear","weekofyear"],train_size)

trainX, trainY = manip.create_dataset_no_look(train)
testX, testY = manip.create_dataset_no_look(test)
```

We used the window method to perform the daily predictions. We took advantage of the time steps recorded in data, and let our model observe Impressions in past days, in order to predict the current and further number of ad impressions. For simplicity, we used only 1 day of previous Impressions. In addition, we chose not to use the DAU values despite its positive effect on the models performance. We would have been required to build another predictor for computing these values.


We trained a simple neural network with 1 input of dimension 4, 1 hidden layer with 15 neurons and an output layer of dimension 1. The model is fitted using mean squared error. We performend a greedy selection of parameters. We tried to minimize the mean squared error by considering a validation set from the training data.

The feature set we used to train our neural network model is:

    1. Previous Impressions. The number of impressions in the previous day
    2. Day of week. The day of the week according to the date
    3. Day of year. The day of the year according to the date
    4. Week of year. The week of the year according to the date
    

```python
# create and fit a Multilayer Perceptron model
model = Sequential()
model.add(Dense(15, input_dim=trainX.shape[1], activation='relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, nb_epoch=200, batch_size=2, verbose=2)
```

The train and test scores of the neural network have a Root mean squared error (RMSE) de 744.27 y 822.21. 

Apart from that, we performed the daily forecast of ad Impressions using the neural network trained.

```python
# Estimate model performance
trainScore = model.evaluate(trainX, trainY, verbose=0)
print('Train Score: %.2f MSE (%.2f RMSE)' % (trainScore, math.sqrt(trainScore)))
testScore = model.evaluate(testX, testY, verbose=0)
print('Test Score: %.2f MSE (%.2f RMSE)' % (testScore, math.sqrt(testScore)))
```

```python
n_testY = np.array([imp_df["Impressions"][train_size]]).astype('float32')
for i in range(1,len(test_df)):
    n_testX = np.array([np.append(test_df[i:i+1].values[0],np.array([n_testY[i-1]]),axis=0)])
    n_testPredict =  model.predict(n_testX[0:1])
    n_testY = np.append(n_testY, n_testPredict[0], axis=0)
```

## 4. Prediction of daily ads Revenue of Game X for February

To predict the daily ads revenue for February, we considered the predicted results of our previous neural network model. Since this model cannot provide a very accurate prediction of the ad Impressions, we performed a very simplistic approach to compute the revenues of this month. We followed the next steps:

    1. We compute the growth rate of daily Impressions from the original data provided
    2. We tuned the predictions of our model with the average growth rate (we assume it is constant through all the time)
    3. We considered we will obtain 0.10 Euros per ad Impression
    4. We computed total income of tuned predictions considering price shown in 3.
    5. We considered eCPM blocks, where a block consisted of 1000 users
    6. We computed eCPM, the cost to reach our users
    7. We obtained the daily revenue by substracting eCPM from the total income 
    
As a result we considered that the average daily revenue for February is 1230.5 Euros with a deviation of +/-27.85 Euros. 

First, we need to consider that this estimation also includes the innacuracy of the Impressions predicted. Additionally, we assumed that the game has a growing average rate of 0.04% per day. We computed an estimated rate of variation between DAU and Impressions of 1.48%, which we used to estimate the DAU per day (since we don't have this information). Finally, we considered we receive 0.10 Euros per ad Impression.

The predictions obtained are saved into the FebDailyRevenue.csv in the predictions folder. However, at the end of this notebook we show the predictions of the first days of February.


```python
df = df[["DAU","Impressions","Date"]]
```

```python
growth_rate = []
for index in df.index.values:
    if index !=0:
        rate = (df["Impressions"][index] - df["Impressions"][index-1])/ df["Impressions"][index-1]
        growth_rate.append(rate)
gr_df = pd.DataFrame(growth_rate,columns=["Growth Rate"])
```

```python
feb_pred["Tunned_Impressions"] = feb_pred["Pred_Impressions"] + round(feb_pred["Pred_Impressions"]*gr_df.mean()[0])

price_impression = 0.15
feb_pred["Total Income"] = feb_pred["Tunned_Impressions"]*price_impression

dau_imp_rate = pd.DataFrame(df["DAU"]/df["Impressions"],columns=["DAU_Imp Rate"])

feb_pred["DAU_estimated"] = feb_pred["Tunned_Impressions"]*dau_imp_rate.mean()[0]

block = 1000
feb_pred["eCPM blocks"] =  feb_pred["DAU_estimated"]/block

feb_pred["eCPM"] = feb_pred["Total Income"]/feb_pred["eCPM blocks"]

feb_pred["Daily Revenue"] = feb_pred["Total Income"] - feb_pred["eCPM"]

Feb_revenue = pd.DataFrame(feb_pred["Daily Revenue"],columns=['Daily Revenue'])
```
